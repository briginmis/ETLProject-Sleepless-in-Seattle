```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import dans_functions as tf

# read in csv
df=pd.read_csv("listings.csv")

# keep only those columns
df = df[['host_id','host_verifications']]

```

```python
df3 = tf.explode_into_rows_in_new_table(input_df = df, id_column = "host_id", explode_column = "host_verifications")
df3
```

```python
# remove the { and } and " from the amenities column
df['amenities']=df['amenities'].str.replace('{','')
df['amenities']=df['amenities'].str.replace('}','')
df['amenities']= df['amenities'].str.replace('"','')

# split the lists inside the amenities column for each row ????
id_list=[]
amenities_list=[]
for index,row in df.iterrows():
    id_list.append(row['id'])
    amenities_list.append(row['amenities'].split(","))
    
# Create a data frame from the id and amenities for each instance
df2=pd.DataFrame({'id':id_list,"amenities":amenities_list})

# Explode (split the lists into new rows)
df2=df2.explode("amenities")
```

```python
# save as csv
df2.to_csv("amenities_clean.csv")

# Preview Output
df2.head(5)
```
