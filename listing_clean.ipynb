```python
def remove_dollar(df,list_of_column_names):
    for names in list_of_column_names:
        df[names]=df[names].str.replace('$','')

def remove_percent(df,list_of_column_names):
    for names in list_of_column_names:
        df[names]=df[names].str.replace('%','')    

def change_true_false(df,list_of_column_names):
    for names in list_of_column_names:
        df[names]=df[names].str.replace('t',True)
        df[names]=df[names].str.replace('f',False)
```
